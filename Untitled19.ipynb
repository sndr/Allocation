```python
import chess
import random

# Reuse the existing ai_move, make_move, and game_status functions
# from previous cells to avoid duplication and keep the code organized.

def play_game_console():
    """Plays a game of chess in the console against the AI."""
    board = chess.Board()
    print("Welcome to Console Chess!")
    print("Enter your moves in UCI format (e.g., e2e4).")
    print(board)

    while not board.is_game_over():
        print("\n" + game_status(board)[1])
        if board.turn == chess.WHITE:
            try:
                move_uci = input("Your move: ")
                success, message = make_move(board, move_uci)
                print(message)
                if success:
                    print(board)
            except EOFError:
                print("\nExiting game.")
                break
        else:
            print("AI is thinking...")
            ai_move_obj = ai_move(board)
            if ai_move_obj:
                board.push(ai_move_obj)
                print(f"AI played: {ai_move_obj.uci()}")
                print(board)
            else:
                print("AI has no legal moves.")
                break

    print("\n" + game_status(board)[1])
    print("Game Over.")

# Start the console game
play_game_console()

# The Tkinter GUI code is commented out as it requires a display environment.
# root = tk.Tk()
# root.title("Chess Game")
# game_board = chess.Board()
# gui = ChessGUI(root, game_board)
# root.mainloop()
```

```python
import chess
import random

def ai_move(board):
    """
    Selects a random legal move for the AI.

    Args:
        board: The current chess board state.

    Returns:
        A chess.Move object representing the chosen move.
    """
    legal_moves = list(board.legal_moves)
    if legal_moves:
        return random.choice(legal_moves)
    return None

def make_move(board, move_uci):
    """Attempts to make a move on the board."""
    try:
        move = board.parse_uci(move_uci)
        # Add print statements to debug why the move is considered illegal
        print(f"Attempting to parse UCI: {move_uci}")
        print(f"Parsed move object: {move}")
        print(f"Is parsed move in board.legal_moves? {move in board.legal_moves}")

        if move in board.legal_moves:
            board.push(move)
            return True, "Move successful."
        else:
            # Add more specific feedback if the move is not legal
            if board.is_check():
                return False, "Illegal move. You are in check."
            elif board.gives_check(move): # This check is tricky with parse_uci, might not work as intended here
                 return False, "Illegal move. Your king would be in check." # Simplified check attempt
            else:
                return False, "Illegal move."
    except ValueError as e:
        print(f"ValueError during move parsing: {e}")
        return False, "Invalid move format."
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return False, "An error occurred during move."


def game_status(board):
    """Checks and reports the current game status."""
    if board.is_game_over():
        result = board.result()
        return True, f"Game over. Result: {result}"
    elif board.is_check():
        return False, "Check!"
    else:
        return False, "Game in progress."

def test_scenario(scenario_name, moves):
    """Tests a specific game scenario with a sequence of moves."""
    print(f"--- Testing Scenario: {scenario_name} ---")
    board = chess.Board()
    print("Initial board:")
    print(board)
    print(game_status(board))

    for i, move_uci in enumerate(moves):
        print(f"\n--- Move {i+1} ---")
        print(f"Attempting move: {move_uci}")

        if board.is_game_over():
            print("Game is already over, skipping move.")
            break

        # Determine whose turn it is for the human move
        is_human_move = (i % 2 == 0) # Assuming human is always white in test scenarios

        if is_human_move and board.turn == chess.WHITE:
            success, message = make_move(board, move_uci)
            print(f"Human move: {message}")
            if success:
                print("Board after human move:")
                print(board)
                print(game_status(board))
            else:
                 print("Board remains unchanged.")

        elif not is_human_move and board.turn == chess.BLACK:
             # This branch is for potential future testing of Black human moves,
             # but in the current setup, White is human and Black is AI.
             # We'll skip this for now and rely on the AI to make Black's moves.
             print("Skipping human move for Black's turn in this test setup.")
             pass # The AI move will handle black's turn

        elif is_human_move and board.turn == chess.BLACK:
             print(f"Skipping human move {move_uci}: It's Black's turn.")

        elif not is_human_move and board.turn == chess.WHITE:
             print(f"Skipping move {move_uci}: It's White's turn, expecting human move.")


        if board.is_game_over():
            print("Game over in scenario.")
            break

        # AI's turn (assuming AI is black)
        if board.turn == chess.BLACK and not board.is_game_over():
             print("AI making a move...")
             ai_move_obj = ai_move(board)
             if ai_move_obj:
                 print(f"AI played: {ai_move_obj.uci()}")
                 board.push(ai_move_obj)
                 print("Board after AI move:")
                 print(board)
                 print(game_status(board))
                 if board.is_game_over():
                     print("Game over after AI move.")
                     break
             else:
                  print("AI has no legal moves.")
                  if board.is_stalemate():
                      print("Game over: Stalemate.")
                  elif board.is_insufficient_material():
                      print("Game over: Insufficient material.")
                  else:
                       print("Game over by no legal moves for AI.") # Should ideally be stalemate or checkmate if no legal moves
                  break


    print(f"--- End of Scenario: {scenario_name} ---")
    print("\n")

# Test Scenarios

# 1. Basic moves and check (corrected sequence)
test_scenario("Basic Moves and Check", ["e2e4", "e7e5", "f1c4", "b8c6", "d1h5", "g8f6"]) # White puts black in check

# 2. Checkmate scenario (Scholar's Mate)
test_scenario("Scholar's Mate Checkmate", ["e2e4", "e7e5", "f1c4", "b8c6", "d1h5", "g8f6", "h5f7"])

# 3. Illegal move attempt
test_scenario("Illegal Move Attempt", ["e2e5"]) # Pawn cannot move 3 squares initially

# 4. Simulate a few moves to observe AI (random) behavior
test_scenario("Observe AI Behavior", ["e2e4", "d2d4", "g1f3", "c2c4"])

# 5. Test Castling (Kingside White)
test_scenario("White Kingside Castling", ["e2e4", "e7e5", "g1f3", "b8c6", "f1e2", "d7d6", "e1g1"])

# 6. Test En Passant (simplified setup)
# Requires a specific board state, so we'll set it up manually for this test
def test_en_passant():
    print("--- Testing Scenario: En Passant ---")
    board = chess.Board("rnbqkbnr/pppppppp/8/8/4P3/8/PPPP1PPP/RNBQKBNR b KQkq - 0 1")
    print("Initial board (setup for en passant):")
    print(board)
    print(game_status(board))

    # Simulate Black's move first to set up the en passant opportunity for White
    success, message = make_move(board, "d7d5")
    print(f"\n--- Move 1 (Simulated Black) ---")
    print(f"Attempting move: d7d5")
    print(f"Simulated Black move: {message}")
    print("Board after simulated Black move:")
    print(board)
    print(game_status(board))


    moves = ["e4d5", "c7c5", "d5c6"] # White captures en passant, Black pawn moves, White attempts another en passant

    for i, move_uci in enumerate(moves):
        print(f"\n--- Move {i+2} ---") # Adjust move number
        print(f"Attempting move: {move_uci}")

        if board.is_game_over():
            print("Game is already over, skipping move.")
            break

        # Alternate turns for human (White) and simulated Black
        if board.turn == chess.WHITE: # White's turn (simulated human)
            success, message = make_move(board, move_uci)
            print(f"Simulated White move: {message}")
            if success:
                print("Board after simulated White move:")
                print(board)
                print(game_status(board))
            else:
                print("Board remains unchanged.")
        else: # Black's turn (simulated opponent/AI - though we are providing moves)
             success, message = make_move(board, move_uci)
             print(f"Simulated Black move: {message}")
             if success:
                 print("Board after simulated Black move:")
                 print(board)
                 print(game_status(board))
             else:
                 print("Board remains unchanged.")


        if board.is_game_over():
            print("Game over in scenario.")
            break

    print(f"--- End of Scenario: En Passant ---")
    print("\n")

test_en_passant()

# 7. Test Pawn Promotion (simplified setup)
def test_pawn_promotion():
    print("--- Testing Scenario: Pawn Promotion ---")
    board = chess.Board("8/P7/8/8/8/8/pppppppp/rnbqkbnr w KQkq - 0 1")
    print("Initial board (setup for pawn promotion):")
    print(board)
    print(game_status(board))

    move_uci = "a7a8q" # Promote pawn to queen

    print(f"\n--- Move 1 ---")
    print(f"Attempting move: {move_uci}")

    if board.is_game_over():
        print("Game is already over, skipping move.")
    elif board.turn == chess.WHITE:
        success, message = make_move(board, move_uci)
        print(f"Simulated White move: {message}")
        if success:
            print("Board after simulated White move:")
            print(board)
            print(game_status(board))
        else:
            print("Board remains unchanged.")
    else:
        print(f"Skipping simulated White move {move_uci}: It's Black's turn.")

    if not board.is_game_over() and board.turn == chess.BLACK:
        print("AI making a move...")
        ai_move_obj = ai_move(board)
        if ai_move_obj:
            print(f"AI played: {ai_move_obj.uci()}")
            board.push(ai_move_obj)
            print("Board after AI move:")
            print(board)
            print(game_status(board))
        else:
            print("AI has no legal moves.")


    print(f"--- End of Scenario: Pawn Promotion ---")
    print("\n")

test_pawn_promotion()

# 8. Test Stalemate (simplified setup)
def test_stalemate():
    print("--- Testing Scenario: Stalemate ---")
    # Corrected FEN for stalemate
    board = chess.Board("8/8/8/8/8/1k6/P7/K7 w - - 0 1") # White to move, no legal moves for White King, not in check
    print("Initial board (setup for stalemate):")
    print(board)
    print(game_status(board))

    if board.is_stalemate():
        print("Board is in a stalemate position as expected.")
    else:
        print("Board is NOT in a stalemate position as expected.")

    print(f"Game Status: {game_status(board)}")


    print(f"--- End of Scenario: Stalemate ---")
    print("\n")

test_stalemate()

# 9. Test Insufficient Material (simplified setup)
def test_insufficient_material():
    print("--- Testing Scenario: Insufficient Material ---")
    # Corrected FEN for King and King
    board = chess.Board("8/8/8/8/8/8/8/k1K5 b - - 0 1") # King and King
    print("Initial board (setup for insufficient material):")
    print(board)
    print(game_status(board))

    if board.is_insufficient_material():
        print("Board has insufficient material as expected.")
    else:
        print("Board does NOT have insufficient material as expected.")

    print(f"Game Status: {game_status(board)}")

    board2 = chess.Board("8/8/8/8/8/8/8/k1KN4 b - - 0 1") # King, Knight, and King
    print("\nInitial board (setup for sufficient material: K+N vs K):")
    print(board2)
    print(game_status(board2))

    if board2.is_insufficient_material():
        print("Board has insufficient material as expected.")
    else:
        print("Board does NOT have insufficient material as expected.")

    print(f"Game Status: {game_status(board2)}")


    print(f"--- End of Scenario: Insufficient Material ---")
    print("\n")

test_insufficient_material()
```
